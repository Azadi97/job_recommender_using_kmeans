### Scrap data including salary

```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.common.exceptions import NoSuchElementException
from time import sleep
from random import randint
from webdriver_manager.chrome import ChromeDriverManager
# from selenium.webdriver.chrome.service import ChromeService
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC

# job_='Data+Engineer'
# location='Ohio'
print(job_, location)

start = time.time()

job_lst = []
salary_list = []

# driver = webdriver.Chrome(service=ChromeService(ChromeDriverManager().install()))
driver = webdriver.Firefox(service=FirefoxService(GeckoDriverManager().install()))
wait = WebDriverWait(driver, 10)

sleep(randint(3, 5))

for i in range(0, max_iter_pgs):
    driver.get(paginaton_url.format(job_, location, i * 10))

    sleep(randint(3, 5))

    job_page = driver.find_element(By.ID, "mosaic-jobResults")
    jobs = job_page.find_elements(By.CLASS_NAME, "job_seen_beacon")  # return a list

    for jj in jobs:
        try:
            job_title = jj.find_element(By.CLASS_NAME, "jobTitle")

            # Close the cookie banner if present
            try:
                cookie_banner = driver.find_element(By.CLASS_NAME, "gnav-CookiePrivacyNoticeBanner")
                cookie_banner.find_element(By.TAG_NAME, "button").click()
                sleep(2)  # Allow time for the banner to close
            except NoSuchElementException:
                pass  # No banner, continue with the click

            # Click the job title to get the description
            job_title.click()

            # Wait for the job description to load
            sleep(randint(3, 5))

            try:
                job_description = wait.until(EC.presence_of_element_located((By.ID, "jobDescriptionText"))).text
            except NoSuchElementException:
                job_description = None

            try:
                # Locate the div with id "salaryInfoAndJobType" and extract salary text
                salary_element = driver.find_element(By.ID, "salaryInfoAndJobType")

                # Extract salary text from the span element inside the div
                salary = salary_element.find_element(By.CSS_SELECTOR, "span.css-19j1a75").text
                print("salary found")

            except NoSuchElementException:
                print("salary not found")
                salary = None

            job_lst.append([
                job_title.text,
                job_title.find_element(By.CSS_SELECTOR, "a").get_attribute("href"),
                job_title.find_element(By.CSS_SELECTOR, "a").get_attribute("id"),
                jj.find_element(By.CSS_SELECTOR, ".company_location [data-testid='company-name']").text,
                jj.find_element(By.CSS_SELECTOR, ".company_location [data-testid='text-location']").text,
                jj.find_element(By.CLASS_NAME, "jobTitle").text,
                jj.find_element(By.CSS_SELECTOR, "[data-testid='myJobsStateDate']").text,
                job_title.find_element(By.CSS_SELECTOR, "a").get_attribute("href"),
                salary
            ])

        except NoSuchElementException:
            job_lst.append(None)

driver.quit()

end = time.time()

print(end - start, 'seconds to complete Query!')
```

```python
columns = ['job_title', 'url', 'job_id', 'company', 'location', 'additional_info', 'posted_date', 'full_url', 'salary']
dfar = pd.DataFrame(job_lst, columns=columns)


dfar.to_csv('washington_ds.csv', index=False)
# pring(len(dfar))
dfar.head(4)
```
